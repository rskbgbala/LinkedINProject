```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
url = "https://www.linkedin.com/jobs/jobs-in-bengaluru?trk=homepage-basic_guest_nav_menu_jobs&position=1&pageNum=0"
#url = "https://www.linkedin.com/jobs/"
html_data = requests.get(url).text
```

```python
soup = BeautifulSoup(html_data,"html.parser")
```

```python
print(soup.prettify())
```

```python
tag_object=soup.find_all("h3")
#tag_object
#tag_object[2].text

company_name=soup.find_all("h4")
#company_name
```

```python
for i,row in enumerate(tag_object):
    if i != 25:
        
        jobtitle = row.text.strip()
        #jobtitle[i]
        print("row",i,"-->", jobtitle)
       
            
        print(company_name[i].text.strip())
             
    print("-------------")   
```
